```python
# Define path to the folder containing your data
folder_path = 'P:/LAB/Manuel/2023/2023-08/Microscopy/20230811_flhE-Dflg-fliCmNG/'
# Load your data - adjust the filename as needed
df = pd.read_csv(folder_path + 'MicrobeJ_Results.csv')

# Your existing data and settings
order = ['EM9357', 'EM13604','EM15552','EM15551','EM15554','EM15553','EM15556','EM15555']
custom_xtitles = ['WT', 'ΔflhE', 'ΔflgBC', 'ΔflgBC ΔflhE', 'ΔflgJ', 'ΔflgJ ΔflhE', 'ΔflgG', 'ΔflgG ΔflhE']

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10,6))

# Plot the 'mNG' data
sns.violinplot(x='IMAGE', y='mNG', data=df, split=True, inner = 'quartile', 
               linewidth=2, color='green', scale='width', cut=False, width=0.8, 
               order=order, ax=ax1)

# Set labels for the first Y-axis
ax1.set_xlabel('Strain')
ax1.set_ylabel('fliC-mNG intensity', fontsize=18)
ax1.set_ylim([0, 10000])  # adjust as needed
ax1.set_xticks(range(len(custom_xtitles)))
ax1.set_xticklabels(custom_xtitles, fontsize=14, rotation=90, style='italic')

# Create a second Y-axis for the 'curvature' data
ax2 = ax1.twinx()

# Plot the 'curvature' data
sns.violinplot(x='IMAGE', y='SHAPE.curvature', data=df, split=True, inner='quartile', 
               linewidth=2, color='darkgray', scale='width', cut=False, width=0.8, 
               order=order, ax=ax2)

# Set labels for the second Y-axis
ax2.set_ylabel('Curvature', fontsize=18)
# Set the limits for the second Y-axis based on your 'curvature' data range
# ax2.set_ylim([min_value, max_value]) 

# Show plot
plt.show()

# Optional: Save figure
# plt.savefig('P:/LAB/Manuel/20230811_flhE-Dflg-fliCmNG/NAME.tif', format='tiff', bbox_inches='tight', dpi=300)
```

```python
def custom_violin(ax, data, pos, fc='b', ec='k', alpha=0.7, percentiles=[25, 50, 75], side="both", scatter_kwargs={}, violin_kwargs={}):
    # Function body - make sure all lines inside the function are indented
    parts = ax.violinplot(data, positions=[pos], **violin_kwargs)
    for pc in parts['bodies']:
        m = np.mean(pc.get_paths()[0].vertices[:, 0])
        points_x = pos
        if side == "left":
            points_x -= 0.05
            pc.get_paths()[0].vertices[:, 0] = np.clip(pc.get_paths()[0].vertices[:, 0], -np.inf, m)
        elif side == "right":
            points_x += 0.05
            pc.get_paths()[0].vertices[:, 0] = np.clip(pc.get_paths()[0].vertices[:, 0], m, np.inf)

        pc.set_facecolor(fc)
        pc.set_edgecolor(ec)
        pc.set_alpha(alpha)

    perc = np.percentile(data, percentiles)
    for p in perc:
        ax.scatter(points_x, p, color=ec, zorder=3, **scatter_kwargs)

# Prepare data - extracting values based on 'order'
order = ['EM9357', 'EM13604', 'EM15552', 'EM15551', 'EM15554', 'EM15553', 'EM15556', 'EM15555']
custom_xtitles = ['WT', 'ΔflhE', 'ΔflgBC', 'ΔflgBC ΔflhE', 'ΔflgJ', 'ΔflgJ ΔflhE', 'ΔflgG', 'ΔflgG ΔflhE']
data1 = [df[df['IMAGE'] == item]['mNG'].values for item in order]
data2 = [df[df['IMAGE'] == item]['SHAPE.curvature'].values for item in order]

s_kwargs = {"s": 40, "marker": "_"}
v_kwargs = {"showextrema": False, "showmedians": False, "showmeans": False, "widths": 0.5}

# create the figure and the axes (left and right)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Plotting the violin plots
for i, (d1, d2) in enumerate(zip(data1, data2)):
    custom_violin(ax1, d1, i - 0.2, '#008000', '#008000', 0.7, scatter_kwargs=s_kwargs, violin_kwargs=v_kwargs)
    custom_violin(ax2, d2, i + 0.2, '#666666', '#666666', 0.7, scatter_kwargs=s_kwargs, violin_kwargs=v_kwargs)

ax1.set_ylabel('Class III $\it{fliC}$-mNG expression', color='#008000', size=20)
ax1.tick_params(axis='y', labelcolor='#008000', size=6, labelsize=16)
ax1.set_ylim(0,10000)
ax2.set_ylabel('Cell curvature', color='#666666', size=20)
ax2.tick_params(axis='y', labelcolor='#666666', size=6, labelsize=16)
ax2.set_ylim(0,0.6)
# set the x-axis tick locations and labels
ax1.set_xticks(range(len(order)))
ax1.set_xticklabels(custom_xtitles, fontsize=16, rotation=45, style='italic',  ha='right')

# Increase the width of the axes and frame
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)  # Adjust the width as needed
    
ax1.set_xlabel('')
ax1.set_title('')
#plt.savefig('Intensity+Curvature_violin-plot.tiff',bbox_inches='tight', dpi=300)
# Show the plot
plt.show()
```

```python
def custom_violin(ax, data, pos, fc='b', ec='k', alpha=0.7, percentiles=[25, 50, 75], side="both", scatter_kwargs={}, violin_kwargs={}):
    parts = ax.violinplot(data, positions=[pos], **violin_kwargs)
    for pc in parts['bodies']:
        m = np.mean(pc.get_paths()[0].vertices[:, 0])
        points_x = pos
        if side == "left":
            points_x -= 0.05
            pc.get_paths()[0].vertices[:, 0] = np.clip(pc.get_paths()[0].vertices[:, 0], -np.inf, m)
        elif side == "right":
            points_x += 0.05
            pc.get_paths()[0].vertices[:, 0] = np.clip(pc.get_paths()[0].vertices[:, 0], m, np.inf)
        pc.set_facecolor(fc)
        pc.set_edgecolor(ec)
        pc.set_alpha(alpha)

    perc = np.percentile(data, percentiles)
    for p in perc:
        ax.scatter(points_x, p, color=ec, zorder=3, **scatter_kwargs)

# Prepare data
order = ['EM9357', 'EM13604', 'EM15552', 'EM15551', 'EM15554', 'EM15553', 'EM15556', 'EM15555']
custom_xtitles = ['WT', 'ΔflhE', 'ΔflgBC', 'ΔflgBC ΔflhE', 'ΔflgJ', 'ΔflgJ ΔflhE', 'ΔflgG', 'ΔflgG ΔflhE']
data1 = [df[df['IMAGE'] == item]['mNG'].values for item in order]
data2 = [df[df['IMAGE'] == item]['SHAPE.curvature'].values for item in order]

s_kwargs = {"s": 40, "marker": "_"}
v_kwargs = {"showextrema": False, "showmedians": False, "showmeans": False, "widths": 0.5}

# Create the figure and the axes (left and right)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Plotting the violin plots
for i, (d1, d2) in enumerate(zip(data1, data2)):
    custom_violin(ax1, d1, i - 0.2, '#008000', '#008000', 0.7, scatter_kwargs=s_kwargs, violin_kwargs=v_kwargs)
    custom_violin(ax2, d2, i + 0.2, '#666666', '#666666', 0.7, scatter_kwargs=s_kwargs, violin_kwargs=v_kwargs)

ax1.set_ylabel('Class III $\it{fliC}$-mNG expression', color='#008000', size=20)
ax1.tick_params(axis='y', labelcolor='#008000', labelsize=16)
ax1.set_ylim(0, 10000)

ax2.set_ylabel('Cell curvature', color='#666666', size=20)
ax2.tick_params(axis='y', labelcolor='#666666', labelsize=16)
ax2.set_ylim(0, 0.6)

# Set the x-axis tick locations and labels
ax1.set_xticks(range(len(order)))
ax1.set_xticklabels(custom_xtitles, fontsize=16, rotation=45, style='italic', ha='right')

# Define and set y-ticks using the axes object for ax1
y_ticks_ax1 = np.arange(0, 10000 + 2000, 2000)  # Adjust the range as needed
y_tick_labels_ax1 = [f"{int(y)}" for y in y_ticks_ax1]  # Format as integers without decimals
ax1.set_yticks(y_ticks_ax1)
ax1.set_yticklabels(y_tick_labels_ax1, fontsize=20)

# Define and set y-ticks using the axes object for ax2
y_ticks_ax2 = np.arange(0, 0.6 + .2, .2)  # Adjust the range as needed
y_tick_labels_ax2 = [f"{y:.1f}" for y in y_ticks_ax2]  # Format to one decimal place
ax2.set_yticks(y_ticks_ax2)
ax2.set_yticklabels(y_tick_labels_ax2, fontsize=20)

# Set minor locator and adjust ticks for ax1 (left y-axis)
ax1.yaxis.set_minor_locator(mticker.MultipleLocator(1000))  # Adjust the locator as needed
ax1.tick_params(axis='y', which='minor', length=5, width=2)  # Adjust width and length of minor ticks
ax1.tick_params(axis='y', which='major', length=10, width=2)  # Adjust width and length of major ticks

# Set minor locator and adjust ticks for ax2 (right y-axis)
ax2.yaxis.set_minor_locator(mticker.MultipleLocator(0.1))  # Adjust the locator as needed
ax2.tick_params(axis='y', which='minor', length=5, width=2)  # Adjust width and length of minor ticks
ax2.tick_params(axis='y', which='major', length=10, width=2)  # Adjust width and length of major ticks

# Adjust tick parameters for the x-axis
ax1.tick_params(axis='x', which='major', length=8, width=2)  

ax1.set_xlabel('')
ax1.set_title('')

# Increase the width of the axes and frame
for spine in ax1.spines.values():
    spine.set_linewidth(2)
for spine in ax2.spines.values():
    spine.set_linewidth(2)
plt.savefig('Intensity+Curvature_violin-plot_Mod.tiff',bbox_inches='tight', dpi=300)
plt.show()
```
